Basic plots with Matplotlib



```
import matplotlib.pyplot as plt
import pandas as pd
```

Line Plot


```
x = pd.Series([1,2,3,5])
y = pd.Series([2,4,1,3])
```

```
plt.plot(x,y)
plt.show()
```

```
x = pd.Series([10,35,15,25,50])
y = pd.Series([1.1,3.5,1.5,2.5,5.0])
```

Scatter Plot

```
fig,ax=plt.subplots()
ax.scatter([x],[y])
plt.show()
```

```
eg = (20, 45, 75, 90, 100, 105, 130, 145)
```

Histogram

```
plt.hist(eg, bins = 6)
plt.show()
```

Line Plot

```
x = ([10, 20, 30, 40])
y = ([20, 40, 10, 30])
x1 = ([10, 20, 30, 40])
y1 = ([40, 10, 30, 20])

plt.plot(x,y,)
plt.plot(x1,y1,'--')

plt.title("Matplotlib")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```
